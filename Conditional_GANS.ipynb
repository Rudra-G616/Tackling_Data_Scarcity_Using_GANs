```python
import requests
from bs4 import BeautifulSoup

def fetch_crystal_system(cod_id):
    url = f'http://www.crystallography.net/cod/{cod_id}.html'
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            parameter_table = soup.find('table', id='entry_data_table')
            parameters = ['a','b','c','α','β','γ']
            p = []
            if parameter_table:
                rows = parameter_table.find_all('tr')
                for row in rows:
                    parameter_name = row.find('th').get_text().strip()
                    parameter_value = row.find('td').get_text().strip()
                    if parameter_name in parameters:
                        p.append(parameter_value)
            return p
        else:
            print(f"Failed to fetch data for COD ID {cod_id}. HTTP status code: {response.status_code}")
            return -1
    except Exception as e:
        print(f"Error: {e}")
        return -1
```

```python
def get_crystal_system(parameters):
    a, b, c, alpha, beta, gamma = parameters
    if a == b == c and alpha == beta == gamma == '90°':
            return 'Cubic'
    elif a == b != c and alpha == beta == gamma == '90°':
            return 'Tetragonal'
    elif a != b != c and alpha == beta == gamma == '90°':
            return 'Orthorhombic'
    elif a == b == c and alpha == beta == gamma != '90°':
            return 'Rhombohedral'
    elif a == b != c and alpha == beta == '90°' and gamma == '120°':
        return 'Hexagonal'
    elif (alpha == gamma == '90°' and beta != '90°' and beta != '120°') or (alpha == beta == '90°' and gamma != '90°' and gamma != '120°') or (alpha == beta == '90°' and gamma != '90°' and gamma != '120°') :
            return 'Monoclinic'
    elif alpha != beta != gamma != '90°':
        return 'Triclinic'
    else:
        return 'Unknown'
```

```python
import pandas as pd
from mendeleev import element as mendeleev_element
from pymatgen.core import Composition

def extract_element_properties(chem_form):
    comp = Composition(chem_form)
    elements = {str(el): comp[el] for el in comp.elements if el.symbol != 'O'}
    if not elements:
            return [None, None, None, None, None, None, None]
    main_element = max(elements, key=elements.get)
    el = mendeleev_element(main_element)
    atomic_radius = el.atomic_radius
    electronegativity = el.electronegativity() if callable(el.electronegativity) else el.electronegativity
    try:
        oxidation_states = el.oxidation_states if not callable(el.oxidation_states) else el.oxidation_states()
        valency = max(oxidation_states, key=abs) if oxidation_states else None
        ionic_radii = el.ionic_radii
        ionic_radius = max(ionic_radii, key=lambda x: abs(x.charge)).ionic_radius if ionic_radii else None
    except Exception as e:
        print(f"An error occurred: {e}")
        valency = None
        ionic_radius = None
    density = el.density
    lattice_constant = el.lattice_constant
    return main_element, atomic_radius, valency, electronegativity, ionic_radius, density, lattice_constant

def enhance_dataset(df):
    df[['Element', 'Atomic Radius', 'Valency',
            'Electronegativity','Ionic Radius', 'Density', 'Lattice Constant']] = df['FormQuimica'].apply(lambda x: pd.Series(extract_element_properties(x)))
    return df

train_data_enhanced = enhance_dataset(train_data)
test_data_enhanced = enhance_dataset(test_data)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models, optimizers
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/enhanced_dataset.csv')

# Drop rows where 'Structure' is NaN or 'Structure' equals 'Unknown'
df = df.dropna(subset=['Structure'])  # This line removes rows where 'Structure' is NaN
df = df[df['Structure'] != 'Unknown']  # This line removes rows where 'Structure' == 'Unknown'

# Separate features and the condition
features = df.drop(columns=['Structure', 'ID', 'Element', 'FormQuimica'], axis=1)
conditions = df[['Structure']]

# Normalize the continuous features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Encode the categorical condition
encoder = OneHotEncoder(sparse_output=False)
conditions_encoded = encoder.fit_transform(conditions)

# Training X and y
X_train = features_scaled
y_train = conditions_encoded

# Create a function to create batches
def get_random_batch(X, y, batch_size):
    """
    Will return random batches of size batch_size

    Params:
    X: numpy array - features
    y: numpy array - classes
    batch_size: Int

    Returns:
    X_batch, y_batch
    """
    idx = np.random.randint(0, len(X))
    X_batch = X[idx:idx+batch_size]
    y_batch = y[idx:idx+batch_size]

    return X_batch, y_batch
```

```python
# Define the generator with varying activation functions
def make_generator_model(activation='relu'):
    if activation == "LeakyReLU":
        activation = layers.LeakyReLU(alpha=0.2)

    noise_input = layers.Input(shape=(24,))
    condition_input = layers.Input(shape=(7,))
    merged_input = layers.Concatenate()([noise_input, condition_input])

    x = layers.Dense(128, use_bias=False)(merged_input)
    x = activation(x) if isinstance(activation, layers.Layer) else layers.Activation(activation)(x)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(256, use_bias=False)(x)
    x = activation(x) if isinstance(activation, layers.Layer) else layers.Activation(activation)(x)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(512, use_bias=False)(x)
    x = activation(x) if isinstance(activation, layers.Layer) else layers.Activation(activation)(x)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(24, activation="tanh")(x)

    model = models.Model([noise_input, condition_input], x)
    return model
```

```python
# Define the discriminator models with varying activation functions
def make_discriminator_model(activation='relu'):
    if activation == "leaky_relu":
        activation = layers.LeakyReLU(alpha=0.2)

    data_input = layers.Input(shape=(24,))
    condition_input = layers.Input(shape=(7,))

    x = layers.Flatten()(data_input)
    merged_input = layers.Concatenate()([x, condition_input])

    x = layers.Dense(512)(merged_input)
    x = activation(x) if isinstance(activation, layers.Layer) else layers.Activation(activation)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Dense(256)(x)
    x = activation(x) if isinstance(activation, layers.Layer) else layers.Activation(activation)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Dense(1)(x)

    model = models.Model([data_input, condition_input], x)
    return model
```

```python
# Define function for training the cGAN with different activation functions
import tensorflow as tf
from tensorflow import keras

def train_cgan(activation_g, activation_d, X_train, y_train, batch_size=128, epochs=100):
    generator = make_generator_model(activation=activation_g)
    discriminator = make_discriminator_model(activation=activation_d)

    discriminator.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0001), loss='binary_crossentropy')
    discriminator.trainable = False

    noise_input = layers.Input(shape=(24,))
    condition_input = layers.Input(shape=(7,))
    gen_data = generator([noise_input, condition_input])

    gan_output = discriminator([gen_data, condition_input])
    gan = keras.Model([noise_input, condition_input], gan_output)
    gan.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0002), loss='binary_crossentropy')

    # Lists to store losses at each epoch
    loss_real_hist = []
    loss_fake_hist = []
    loss_gan_hist = []

    for epoch in range(epochs):
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        X_batch, labels = X_train[idx], y_train[idx]

        # Train with real values
        y_real = np.ones((batch_size, 1))
        loss_real = discriminator.train_on_batch([X_batch, labels], y_real)
        loss_real_hist.append(loss_real)

        # Generate fake data and train with it
        noise = np.random.normal(0, 1, (batch_size, 24))
        X_fake = generator.predict([noise, labels])
        y_fake = np.zeros((batch_size, 1))
        loss_fake = discriminator.train_on_batch([X_fake, labels], y_fake)
        loss_fake_hist.append(loss_fake)

        # Train generator via GAN
        y_gan = np.ones((batch_size, 1))
        loss_gan = gan.train_on_batch([noise, labels], y_gan)
        loss_gan_hist.append(loss_gan)

    return loss_real_hist, loss_fake_hist, loss_gan_hist
```

```python
# Perform hyperparameter search by training cGAN with different activation functions
from tensorflow.keras.layers import LeakyReLU

activation_functions = ['relu', 'tanh', 'sigmoid', LeakyReLU(alpha=0.2)]  # Use LeakyReLU as a layer
final_losses = {}

for activation_g in activation_functions:
    for activation_d in activation_functions:
        key = f'{activation_g}-{activation_d}' if isinstance(activation_g, str) else 'LeakyReLU-LeakyReLU'
        print(f"Training cGAN with Generator Activation: {activation_g}, Discriminator Activation: {activation_d}")

        # Ensure LeakyReLU is passed correctly
        generator_activation = activation_g if isinstance(activation_g, str) else LeakyReLU(alpha=0.2)
        discriminator_activation = activation_d if isinstance(activation_d, str) else LeakyReLU(alpha=0.2)

        loss_real_hist, loss_fake_hist, loss_gan_hist = train_cgan(generator_activation, discriminator_activation, X_train, y_train)

        # Store final loss for this combination of activation functions
        final_losses[key] = loss_gan_hist[-1]

# Plot the final Loss with different activation functions
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(final_losses.keys(), final_losses.values())
plt.xlabel('Activation Function Combinations')
plt.ylabel('Final Loss')
plt.title('Final Loss with Different Activation Function Combinations')
plt.xticks(rotation=45)
plt.savefig('/content/final_loss_plot.png')
plt.show()
```

```python
# Finding number of epochs necessary
loss_real_hist_f, loss_fake_hist_f, loss_gan_hist_f = train_cgan('relu', 'relu', X_train, y_train)

# Ensure x_epochs matches the length of the loss arrays
x_epochs = range(1, len(loss_real_hist_f) + 1)

plt.figure(figsize=(10, 6))

# Plot loss versus epochs for discriminator trained on real data
plt.plot(x_epochs, loss_real_hist_f, label='Discriminator (Real)', color='blue')

# Plot loss versus epochs for discriminator trained on fake data
plt.plot(x_epochs, loss_fake_hist_f, label='Discriminator (Fake)', color='green')

# Plot loss versus epochs for generator trained via GAN
plt.plot(x_epochs, loss_gan_hist_f, label='Generator (GAN)', color='red')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs. Epochs for cGAN (Sigmoid Generator, ReLU Discriminator)')
plt.legend()
plt.grid(True)
plt.savefig('/content/final_loss_vs_epoch.png')
plt.show()

```

```python
# Final training function based on above decisions
def cgan(X_train, y_train):
    batch_size = 128
    epochs = 100

    # Create generator and discriminator models with specified activation
    generator = make_generator_model(activation='relu')
    discriminator = make_discriminator_model(activation='relu')

    # Compile discriminator model
    discriminator.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0002), loss = 'binary_crossentropy')
    discriminator.trainable = False

    # Input layers for noise and condition
    noise_input = layers.Input(shape=(24,))
    condition_input = layers.Input(shape=(7,))

    # Generate data using generator
    gen_data = generator([noise_input, condition_input])

    # Pass generated data and condition to discriminator
    gan_output = discriminator([gen_data, condition_input])

    # Combine generator and discriminator into GAN
    gan = keras.Model([noise_input, condition_input], gan_output)
    gan.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0002), loss='binary_crossentropy')

    # Lists to store loss history
    loss_real_hist = []
    loss_fake_hist = []
    loss_gan_hist = []

    for epoch in range(epochs):
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        X_batch, labels = X_train[idx], y_train[idx]

        # Train discriminator with real values
        y_real = np.ones((batch_size, 1))
        loss_real = discriminator.train_on_batch([X_batch, labels], y_real)
        loss_real_hist.append(loss_real)

        # Generate fake data and train discriminator with it
        noise = np.random.normal(0, 1, (batch_size, 24))
        X_fake = generator.predict([noise, labels])
        y_fake = np.zeros((batch_size, 1))
        loss_fake = discriminator.train_on_batch([X_fake, labels], y_fake)
        loss_fake_hist.append(loss_fake)

        # Train generator via GAN
        y_gan = np.ones((batch_size, 1))
        loss_gan = gan.train_on_batch([noise, labels], y_gan)
        loss_gan_hist.append(loss_gan)

    return loss_real_hist, loss_fake_hist, loss_gan_hist, generator
```

```python
# Generate synthetic data using the trained generator
def generate_synthetic_data(generator, scaler, encoder, num_samples=1000):
    synthetic_noise = np.random.normal(0, 1, size=(num_samples, 24))
    synthetic_conditions = np.eye(7)[np.random.randint(0, 7, size=num_samples)]
    synthetic_data = generator.predict([synthetic_noise, synthetic_conditions])
    synthetic_features = scaler.inverse_transform(synthetic_data[:, :24])
    synthetic_structure = encoder.inverse_transform(synthetic_conditions)
    synthetic_df = pd.DataFrame(np.hstack((synthetic_features, synthetic_structure)))

    return synthetic_df

_, _, _, generator = cgan(X_train, y_train)

synthetic_df_1000 = generate_synthetic_data(generator, scaler, encoder, num_samples=1000)
augmented_df_1000 = pd.concat([df, synthetic_df_1000], ignore_index=True)
print("Shape of augmented dataset with 1000 additional samples:", augmented_df_1000.shape)

synthetic_df_5000 = generate_synthetic_data(generator, scaler, encoder, num_samples=5000)
augmented_df_5000 = pd.concat([df, synthetic_df_5000], ignore_index=True)
print("Shape of augmented dataset with 5000 additional samples:", augmented_df_5000.shape)

synthetic_df_10000 = generate_synthetic_data(generator, scaler, encoder, num_samples=10000)
augmented_df_10000 = pd.concat([df, synthetic_df_10000], ignore_index=True)
print("Shape of augmented dataset with 10000 additional samples:", augmented_df_10000.shape)

# Save augmented DataFrame with 1000 additional samples to CSV
augmented_df_1000.to_csv('/content/augmented_data_1000.csv')

# Save augmented DataFrame with 5000 additional samples to CSV
augmented_df_5000.to_csv('/content/augmented_data_5000.csv')

# Save augmented DataFrame with 10000 additional samples to CSV
augmented_df_10000.to_csv('/content/augmented_data_10000.csv')


```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import Callback
from sklearn.preprocessing import StandardScaler, LabelEncoder
from minisom import MiniSom
import tensorflow.keras.backend as K
```

```python
# Cyclical Learning Rate (CLR) Callback Definition
class CyclicLR(Callback):
    # Constructor to initialize the cyclical learning rate parameters
    def __init__(self, base_lr=0.001, max_lr=0.006, step_size=2000,
                 scale_fn=None, scale_mode='cycle'):
        super(CyclicLR, self).__init__()  # Call the constructor of the parent class 'Callback'
        self.base_lr = base_lr  # The initial learning rate which is the lower boundary in the cycle
        self.max_lr = max_lr    # The maximum learning rate which is the upper boundary in the cycle
        self.step_size = step_size  # Total number of steps (batches) per half-cycle
        self.scale_fn = scale_fn or (lambda x: 1.)  # Custom scaling function for the learning rate
        self.scale_mode = scale_mode  # 'cycle' indicates that scale_fn is evaluated on cycle number
        self.clr_iterations = 0.  # Counter for the number of iterations

    # Method to calculate the learning rate at batch end
    def clr(self):
        cycle = np.floor(1 + self.clr_iterations / (2 * self.step_size))  # Compute the current cycle number
        x = np.abs(self.clr_iterations / self.step_size - 2 * cycle + 1)  # Position within the cycle
        return self.base_lr + (self.max_lr - self.base_lr) * np.maximum(0, (1 - x)) * self.scale_fn(cycle)

    # Called once before the model starts training
    def on_train_begin(self, logs=None):
        logs = logs or {}
        # Set the initial learning rate before training starts
        if self.clr_iterations == 0:
            # K.set_value(self.model.optimizer.learning_rate, self.base_lr)
            self.model.optimizer.learning_rate.assign(self.base_lr)

        else:
            # K.set_value(self.model.optimizer.learning_rate, self.clr())
            self.model.optimizer.learning_rate.assign(self.clr())

    # Called at the end of each batch processing
    def on_batch_end(self, epoch, logs=None):
        logs = logs or {}
        self.clr_iterations += 1  # Increment the total number of iterations
        # K.set_value(self.model.optimizer.lr, self.clr())  # Update the learning rate
        self.model.optimizer.learning_rate.assign(self.clr())
```

```python
def determine_optimal_epoch_with_SOM(X, max_epochs=5000, som_grid_size=(10, 10), sigma=1.0, learning_rate=0.5, window_size=50):
    # Initialize a MiniSom instance with specified grid size, input dimensions,
    # neighborhood function width (sigma), and learning rate. This sets up the SOM with a grid of 10x10 neurons,
    # each corresponding to input features.
    som = MiniSom(som_grid_size[0], som_grid_size[1], X.shape[1], sigma=sigma, learning_rate=learning_rate)

    quantization_errors = []  # List to store the quantization errors for each epoch

    # Train the SOM for a specified number of epochs; each epoch involves a
    # random training instance from the dataset.
    for i in range(1, max_epochs + 1):
        som.train_random(X, num_iteration=1)  # Random training improves the generalization capability of the SOM.
        quantization_errors.append(som.quantization_error(X))  # Calculate and store the quantization error.

    # Smooth the quantization error values using a moving average (sliding window) to reduce noise and fluctuations.
    window_avg = np.convolve(quantization_errors, np.ones(window_size) / window_size, mode='valid')

    # Identify the point where the rate of change (derivative) of the smoothed quantization error is minimized,
    # which suggests a balance between learning and overfitting.
    changes = np.abs(np.diff(window_avg))
    optimal_epoch = np.argmin(changes) + (window_size // 2)  # Adjust index due to windowing effect.

    return optimal_epoch  # Return the epoch number where the quantization error stabilizes.
```

```python
# Load and preprocess the data
train_df = pd.read_csv('/content/enhanced_dataset.csv')  # Load the training dataset from a CSV file.
test_df = pd.read_csv('/content/enhanced_test_dataset.csv')    # Load the test dataset from a CSV file.

# Data Cleaning:
train_df = train_df.drop(['FormQuimica', 'ID', 'Element'], axis=1)
# Remove irrelevant or unnecessary columns from the training data.
train_df = train_df[train_df['Structure'] != 'unknown']
# Filter out rows where the structure is unknown.

test_df = test_df.drop(['FormQuimica', 'ID', 'Element'], axis=1)
# Remove the same columns from the test data.
test_df = test_df[test_df['Structure'] != 'unknown']
# Filter out rows where the structure is unknown in the test data.

# Handling Missing Values:
numeric_columns = train_df.select_dtypes(include=[np.number]).columns
# Identify numerical columns in the training data.

train_df[numeric_columns] = train_df[numeric_columns].fillna(train_df[numeric_columns].mean())
# Fill missing values in numerical columns with the mean of each column in the training data.

test_df[numeric_columns] = test_df[numeric_columns].fillna(test_df[numeric_columns].mean())
# Similarly, fill missing values in numerical columns in the test data.

# Encoding Categorical Data:
structure_encoder = LabelEncoder()
# Initialize a label encoder for the 'Structure' categorical column.

train_df['Structure'] = structure_encoder.fit_transform(train_df['Structure'])
# Fit the encoder and transform 'Structure' to numeric labels in the training data.

test_df['Structure'] = structure_encoder.transform(test_df['Structure'])
# Transform 'Structure' to numeric labels in the test data using the same encoder.

# Feature Scaling:
X_train = train_df.drop('Structure', axis=1) # Separate features from labels in the training data.
y_train = train_df['Structure'] # Separate labels for training data.
X_test = test_df.drop('Structure', axis=1) # Separate features from labels in the test data.
y_test = test_df['Structure'] # Separate labels for test data.

scaler = StandardScaler() # Initialize a standard scaler to normalize the feature data.
X_train_scaled = scaler.fit_transform(X_train) # Fit the scaler on the training features and transform them.
X_test_scaled = scaler.transform(X_test) # Transform the test features using the same scaler.

```

```python
# Determine optimal epoch using SOM
optimal_epoch = determine_optimal_epoch_with_SOM(X_train_scaled)
print(f"Optimal number of epochs determined by SOM: {optimal_epoch}")
```

```python
# Function to train and evaluate the model with different activation functions
def train_and_evaluate_model(activation_function):
    # Initialize a sequential model
    model = Sequential()

    # Add a dense layer with 128 neurons and input shape matching the feature set
    model.add(Dense(128, input_shape=(X_train_scaled.shape[1],)))

    # Conditionally add activation layers based on the specified activation function
    if activation_function == 'leaky_relu':
        # Use LeakyReLU for handling negative inputs more effectively
        model.add(LeakyReLU(alpha=0.1))
    else:
        # Use traditional activation functions such as relu, sigmoid, or tanh
        model.add(Activation(activation_function))

    # Add another dense layer with 64 neurons
    model.add(Dense(64))

    # Repeat the activation choice logic
    if activation_function == 'leaky_relu':
        model.add(LeakyReLU(alpha=0.1))
    else:
        # Use traditional activation functions such as relu, sigmoid, or tanh
        model.add(Activation(activation_function))

    # Add another dense layer with 32 neurons
    model.add(Dense(32))

    # Repeat the activation choice logic
    if activation_function == 'leaky_relu':
        model.add(LeakyReLU(alpha=0.1))
    else:
        model.add(Activation(activation_function))

    # Add the output layer with a softmax activation to classify into multiple classes
    model.add(Dense(len(structure_encoder.classes_), activation='softmax'))

    # Define the optimizer (Adam) and the learning rate adjustment callback (CyclicLR)
    optimizer = Adam()
    clr = CyclicLR(base_lr=0.001, max_lr=0.006, step_size=2000.)

    # Compile the model with specified loss function and accuracy as the metric
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    # Train the model using the training data, validation split for tuning, and the CLR callback
    history = model.fit(X_train_scaled, y_train, epochs=optimal_epoch, batch_size=256, validation_split=0.2, callbacks=[clr], verbose=1)

    # Evaluate the model on the test set
    test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test, verbose=0)

    # Output the test accuracy for each activation function
    print(f"{activation_function} - Test Accuracy: {test_accuracy*100:.2f}%")

    # Return the test accuracy and training history for further analysis
    return test_accuracy, history
```

```python
# List of activation functions to be tested
activation_functions = ['relu', 'sigmoid', 'tanh', 'leaky_relu']

# Dictionaries to store training histories and accuracies
model_histories = {}
model_accuracies = {}

# Loop through each activation function, train and evaluate the model, and store the results
for activation in activation_functions:
    accuracy, history = train_and_evaluate_model(activation)
    model_histories[activation] = history
    model_accuracies[activation] = accuracy
```

```python
# Import matplotlib's pyplot to create plots
import matplotlib.pyplot as plt

# Loop through each activation function to plot their respective accuracies
for activation in activation_functions:
    # Retrieve the training history for the current activation function
    history = model_histories[activation]
    # Retrieve the test accuracy for the current activation function
    accuracy = model_accuracies[activation]

    # Create a new figure for the plot with a specified size
    plt.figure(figsize=(10, 5))
    # Plot the training accuracy history
    plt.plot(history.history['accuracy'], label=f'Training accuracy ({activation})')
    # Plot the validation accuracy history
    plt.plot(history.history['val_accuracy'], label=f'Validation accuracy ({activation})')
    # Add a title to the plot including the test accuracy
    plt.title(f'Training and Validation Accuracy with {activation}\nTest Accuracy: {accuracy*100:.2f}%')
    # Label the x-axis as 'Epochs'
    plt.xlabel('Epochs')
    # Label the y-axis as 'Accuracy'
    plt.ylabel('Accuracy')
    # Add a legend to the plot to identify which line is which
    plt.legend()
    # Display the plot
    plt.show()
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import Callback, EarlyStopping
from sklearn.preprocessing import StandardScaler, LabelEncoder
from minisom import MiniSom
import tensorflow.keras.backend as K
import matplotlib.pyplot as plt

# Cyclical Learning Rate (CLR) Callback Definition
class CyclicLR(Callback):
    def __init__(self, base_lr=0.001, max_lr=0.006, step_size=2000., scale_fn=None, scale_mode='cycle'):
        super(CyclicLR, self).__init__()
        self.base_lr = base_lr
        self.max_lr = max_lr
        self.step_size = step_size
        self.scale_fn = scale_fn or (lambda x: 1.)
        self.scale_mode = scale_mode
        self.clr_iterations = 0.

    def clr(self):
        cycle = np.floor(1 + self.clr_iterations / (2 * self.step_size))
        x = np.abs(self.clr_iterations / self.step_size - 2 * cycle + 1)
        return self.base_lr + (self.max_lr - self.base_lr) * np.maximum(0, (1 - x)) * self.scale_fn(cycle)

    def on_train_begin(self, logs=None):
        logs = logs or {}
        if self.clr_iterations == 0:
            self.model.optimizer.learning_rate.assign(self.base_lr)
        else:
            self.model.optimizer.learning_rate.assign(self.clr())

    def on_batch_end(self, batch, logs=None):
        logs = logs or {}
        self.clr_iterations += 1
        self.model.optimizer.learning_rate.assign(self.clr())

def determine_optimal_epoch_with_SOM(X, max_epochs=5000, som_grid_size=(10, 10), sigma=1.0, learning_rate=0.5, window_size=50):
    som = MiniSom(som_grid_size[0], som_grid_size[1], X.shape[1], sigma=sigma, learning_rate=learning_rate)

    quantization_errors = []

    for i in range(1, max_epochs + 1):
        som.train_random(X, num_iteration=1)
        quantization_errors.append(som.quantization_error(X))

    window_avg = np.convolve(quantization_errors, np.ones(window_size) / window_size, mode='valid')

    changes = np.abs(np.diff(window_avg))
    optimal_epoch = np.argmin(changes) + (window_size // 2)

    return optimal_epoch

# Load and preprocess the data
train_df = pd.read_csv('/content/augmented_data_1000.csv', index_col=0)
test_df = pd.read_csv('/content/enhanced_test_dataset.csv', index_col=0)

print("Train columns:", train_df.columns)
print("Test columns:", test_df.columns)

# Remove rows where 'Structure' is NaN or 'unknown'/'Unknown'
train_df = train_df.dropna(subset=['Structure'])
train_df = train_df[~train_df['Structure'].str.lower().isin(['unknown'])]

# Drop unnecessary columns
columns_to_drop = ['FormQuimica', 'ID', 'Element']
test_df = test_df.drop(columns=[col for col in columns_to_drop if col in test_df.columns], errors='ignore')
train_df = train_df.loc[:, test_df.columns]
test_df = test_df[~test_df['Structure'].str.lower().isin(['unknown'])]

if 'Structure' not in train_df.columns or 'Structure' not in test_df.columns:
    raise ValueError("'Structure' column not found in the dataset")

# Handle numeric columns
numeric_columns = train_df.select_dtypes(include=[np.number]).columns
train_df[numeric_columns] = train_df[numeric_columns].fillna(train_df[numeric_columns].mean())

# For test_df, only use columns that exist in both dataframes
common_columns = test_df.columns.intersection(train_df.columns)
test_df = test_df[common_columns]
numeric_columns = test_df.select_dtypes(include=[np.number]).columns
test_df[numeric_columns] = test_df[numeric_columns].fillna(test_df[numeric_columns].mean())

structure_encoder = LabelEncoder()
train_df['Structure'] = structure_encoder.fit_transform(train_df['Structure'])
test_df['Structure'] = structure_encoder.transform(test_df['Structure'])

X_train = train_df.drop('Structure', axis=1)
y_train = train_df['Structure']
X_test = test_df.drop('Structure', axis=1)
y_test = test_df['Structure']

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Determine optimal epoch using SOM
optimal_epoch = determine_optimal_epoch_with_SOM(X_train_scaled)
print(f"Optimal number of epochs determined by SOM: {optimal_epoch}")

def train_and_evaluate_model(activation_function):
    model = Sequential()
    model.add(Dense(128, input_shape=(X_train_scaled.shape[1],)))
    if activation_function == 'leaky_relu':
        model.add(LeakyReLU(alpha=0.1))
    else:
        model.add(Activation(activation_function))
    model.add(Dense(64))
    if activation_function == 'leaky_relu':
        model.add(LeakyReLU(alpha=0.1))
    else:
        model.add(Activation(activation_function))
    model.add(Dense(32))
    if activation_function == 'leaky_relu':
        model.add(LeakyReLU(alpha=0.1))
    else:
        model.add(Activation(activation_function))
    model.add(Dense(len(structure_encoder.classes_), activation='softmax'))

    optimizer = Adam()
    clr = CyclicLR(base_lr=0.001, max_lr=0.006, step_size=2000.)

    # Early stopping callback
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        mode='min',
        restore_best_weights=True
    )

    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    history = model.fit(
        X_train_scaled, y_train,
        epochs=optimal_epoch,
        batch_size=256,
        validation_split=0.2,
        callbacks=[clr, early_stopping],
        verbose=1
    )

    test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test, verbose=0)

    print(f"{activation_function} - Test Accuracy: {test_accuracy*100:.2f}%")

    return test_accuracy, history

activation_functions = ['relu', 'sigmoid', 'tanh', 'leaky_relu']

model_histories = {}
model_accuracies = {}

for activation in activation_functions:
    accuracy, history = train_and_evaluate_model(activation)
    model_histories[activation] = history
    model_accuracies[activation] = accuracy

# Plotting results
for activation in activation_functions:
    history = model_histories[activation]
    accuracy = model_accuracies[activation]

    plt.figure(figsize=(10, 5))
    plt.plot(history.history['accuracy'], label='Training accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation accuracy')
    plt.title(f'Training and Validation Accuracy with {activation}\nTest Accuracy: {accuracy*100:.2f}%')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```
